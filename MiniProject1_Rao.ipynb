# Data wrangling and Data cleaning

```python
#Dropping insignificant columns
airbnbData.drop(['id','host_name','last_review'], axis=1, inplace=True)
```

```python
#Replacing null values with 0
airbnbData.fillna({'reviews_per_month':0}, inplace = True)
```

```python
#Checking to see if null values exist
airbnbData.reviews_per_month.isnull().sum()
```

# Exploratory Data Analysis on the dataset



```python
#Finding hosts with the most listings
MaxHostListings = airbnbData.host_id.value_counts()
```

```python
#Top 10 hosts with most listings
MaxHostListings = MaxHostListings.head(10)
```

```python
#Printing it as structured data
TopHosts = pd.DataFrame(MaxHostListings)
TopHosts.reset_index(inplace=True)
TopHosts.rename(columns={'index':'HostID', 'host_id':'Count of Listings with the HostID'}, inplace=True)
TopHosts
```

```python
#10 most reviewed listings 
MostReviewedListings = airbnbData.nlargest(10,'number_of_reviews')
MostReviewedListings
```

```python
#Summary table of neighbourhood groups & room type  against number of reviews
pd.DataFrame(airbnbData.groupby(["neighbourhood_group","room_type"])["number_of_reviews"].count())
```
